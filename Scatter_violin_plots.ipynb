```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
from itertools import combinations

# Assuming 'df' is your DataFrame with the data already loaded

# Define the list of variables you're interested in
variables = ['SiO2', 'Al2O3', 'Fe2O3', 'K2O', 'TiO2', 'Th', 'CaO']

# Calculate all unique pairs of variables for joint plots
variable_pairs = list(combinations(variables, 2))

# Choose a palette with more distinct colors
palette = 'tab10'  # This palette has distinct colors for better discrimination

# Define marker styles for different 'Environment' categories
markers = ["o", "s", "^"]  # Circle, Square, Triangle

# Plotting each pair with scatter plot with hue based on 'Environment'
for pair in variable_pairs:
    # Create the scatter plot
    plt.figure(figsize=(8, 6))

    # Create a scatter plot for each 'Environment' category with different markers
    for i, category in enumerate(df['Environment'].unique()):
        category_df = df[df['Environment'] == category]
        sns.scatterplot(x=category_df[pair[0]], y=category_df[pair[1]],
                        label=category, marker=markers[i % len(markers)], palette=palette)

    # Calculate the regression line properties for all points, ignoring the hue
    slope, intercept, r_value, p_value, std_err = linregress(df[pair[0]], df[pair[1]])

    # Add the best fit line for all points to the scatter plot
    plt.plot(df[pair[0]], intercept + slope * df[pair[0]], color='red', lw=2)

    # Annotate the plot with the R-squared value
    plt.text(0.05, 0.95, f'$R^2 = {r_value ** 2:.2f}$', transform=plt.gca().transAxes,
             verticalalignment='top', color='red', fontsize=12, bbox=dict(facecolor='white', edgecolor='black'))

    # Set the axis limits based on the data
    plt.xlim(df[pair[0]].min(), df[pair[0]].max())
    plt.ylim(df[pair[1]].min(), df[pair[1]].max())

    # Add grid to the plot
    plt.grid(True)

    # Remove the legend from the current plot to avoid duplication in each loop
    plt.legend(title='Environment', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    # Adjust the figure layout to prevent overlap of the legend and plot
    plt.tight_layout(rect=[0, 0, 0.75, 1])

    plt.show()

# Note: The separate figure for the unified legend is omitted here, assuming it's not needed anymore.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Define feature groups
features_group1 = ['SiO2', 'Al2O3', 'Fe2O3', 'K2O', 'TiO2', 'Th', 'CaO', 'C',  'Zr']
features_group2 = ['MgO', 'Na2O', 'MnO', 'P2O5', 'SrO', 'S', 'Ba', 'Sr', 'U']

def plot_violin_plots(features, group_name):
    plt.figure(figsize=(16, 10))
    for i, feature in enumerate(features):
        plt.subplot(2, 5, i+1)
        sns.violinplot(x='Environment', y=feature, data=df, palette='Set1')
        plt.title(feature)
        plt.xlabel('')
        plt.ylabel('')
    plt.suptitle(f'Violin Plots for {group_name}', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# Plot violin plots for the first group of features
plot_violin_plots(features_group1, 'Group 1 Features')

# Plot violin plots for the second group of features
plot_violin_plots(features_group2, 'Group 2 Features')
```
