```python
file_name_1 = "New Carbonate environment-3-classes.xlsx"
df = pd.read_excel(file_name_1)
```

```python
#Correlation-coefficient-heatmap
columns_of_interest = ['SiO2', 'Al2O3', 'Fe2O3', 'K2O', 'TiO2', 'Th', 'CaO', 'MgO', 'C',
       'Na2O', 'MnO', 'P2O5', 'SrO', 'S', 'Ba', 'Sr', 'U', 'V', 'Zr', 'Zn',
       'Ce ']  # Adjust your feature list
df_selected = df[columns_of_interest]
df_imputed = df_selected.fillna(df_selected.mean())

# Calculate correlation matrix
corr_matrix = df_imputed.corr()
corr_matrix = df_imputed.corr()

# Generate a clustermap
sns.clustermap(corr_matrix, figsize=(15, 15), annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5, annot_kws={"size": 12})
plt.title('Hierarchical Clustering of Features Based on Correlation')
plt.show()
```
